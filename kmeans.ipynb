```python
import numpy as np
from matplotlib import pyplot as plt
import mglearn
import pandas as pd
data = pd.read_csv('DatosVPI_KAJA/DatosVPI/28Gbd_16QAM/D1_LW1_LP100e-3/D1_LW1_LP100e-3.csv')
data
```

```python
from sklearn.model_selection import train_test_split

X = data.values[:,:2]
print(X)
#y = data.values[:,2]
#X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, shuffle = False)
X_train, X_test= train_test_split(X, test_size = 0.3, shuffle = False)
plt.scatter(X_train[:,0], X_train[:,1])
plt.scatter(X_train[:,0], X_train[:,1])
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=16)
kmeans.fit(X_train)

assignments = kmeans.labels_
mglearn.discrete_scatter(X_train[:, 0], X_train[:, 1], assignments)
```

```python
y_pred = kmeans.predict(X_train)
if False in (kmeans.labels_ == y_pred).tolist(): print("One error") 
else: print("No error")
errors = (kmeans.labels_ != y_pred).sum()
score = errors/len(y_pred)
print("The score is",(1-score))
```

```python
import numpy as np
from matplotlib import pyplot as plt
import mglearn
import pandas as pd
data = pd.read_csv('DatosVPI_KAJA/DatosVPI/28Gbd_16QAM/D1_LW25K_LP100e-3/D1_LW25K_LP100e-3.csv')
data
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=16)
kmeans.fit(X_train)

assignments = kmeans.labels_
mglearn.discrete_scatter(X_train[:, 0], X_train[:, 1], assignments)
```
